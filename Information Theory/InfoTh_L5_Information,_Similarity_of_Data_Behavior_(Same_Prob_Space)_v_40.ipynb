#### Notes and  Notations

- Here I use  "Domain" and "Space" interchangeably, as used in "Sample Domain/Space" and "Probability Domain/Space" (see Fig. 1-4)

- When speaking with statisticians, the words as "system" or "experiments" should be replaced by term "random process", and word "data collection" or "measurement" or "recordings" should be replaced by "sample selection" or "observation".</small>
<br>

- It is common, e.g. also in [1], to denote $X$ be a random variable and $x$ be its value, so the notations would yield $p(X=x)$. Bellow, a scalar value of random variable $X$ is denoted with its sample index as $x(k)$ as for time series, or just without the index $(k)$ as for histograms, see the following text, codes, and graphs for Fig.1-Fig.3.

- Also, small $p$ is usually used for values of probability mass function, while capital  $P$ is used for probability in theoretical derivations.

Therefore, notation for random variables, vs. matrix, vector (1-D array), or a scalar is as follows:

- $X,Y$ ... random variables, capital italic (not bold),
- $\textbf{X},\textbf{Y}$ ... matrix, bold capital, where particularly $\textbf{X}$ can be for a matrix of row feature vectors as inputs to a neural network  
- $\textbf{x},\textbf{y}$ ... a vector (1-D array), small and bold, it can be either the whole set of values as 1-D array (data observation); in another way (a bit confusingly), letter "x" as $\textbf{x}$  also denotes a feature vector as for a neural network; $\textbf{y}$ is then network output or labels (targets).
- $x, y, x(k), y(k)$ ... scalars, small and italic, particular values, i.e., selected sample (value) of random variables $X$, $Y$

- Curly bracketes (braces) $\{ \ \}$ for $x$ such as $\{ x \} $ usually denotes an array of values such as $\{ x \text{ for } k=1,2,..,N \}=[x(1)  \ , \  x(2) \ , \ ...\ x(N)] $. For the meaning of "ensemble", see [1] (subsection 2.1., p.22 and futher),

- $p_x,p_y$ ... probabilities, i.e., probability mass functions of discrete random variables (e.g. as obtained from histograms, see Fig.1-Fig.3)

- $P$ ... probability in general sense (in theoretical formulas)

<h3> Sample Domain, Histogram, and the Same Probability Space </h3>

Let's have two data observations (for simplicity two 1-D arrays) as follows <br>

<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\textbf{x}=[\{ x(k) \text{ for } k=1,2,3,...,N_x\}] \text{ where } \ x\in \langle x_{min}, x_{max} \rangle, $<br>
    $\textbf{y}=[\{ y(k) \text{ for } k=1,2,3,...,N_y\}] \text{ where } \ y\in \langle y_{min}, y_{max} \rangle ,$ </div>
    <div style="width:10%;height:40px;float:left;text-align:right">(1)</div>
</div>

where $\textbf{x}$ and $\textbf{y}$ are observations of random variables $X$ and $Y$, $k$ stands for sample index and $N_x, N_y$ denote the data lengths. So, $x(k)$ and $y(k)$ represent observed data from two different systems or these can be data colections (from two different recordings at the same system, at different times, or by two different sensors , etc...).




<center><b>Fig. 1</b>: Sample values $x(k)$ and $y(k)$ of random variables $X$ and $Y$ in their sample domains.</center>

The (discrete) probability of $x$, i.e., probability mass function of discrete random variable  $x$, can be here denoted as $p_x(x)$ that means a 1-D array as follows
<br>

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $\textbf{p}_x=[{p_x(x) \ \forall \ x}] = [p_x(x_1) \ , \ p_x(x_2) \ ,\dots ,  \ p_x(x_i) \ , \dots \ ,\ p_x(x_{n_{bins}})],$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(2)</div>
</div>


<br>
where $x_i$ are bin centers for $i=1,2,\dots,n_{bins}$ , so $\textbf{x}$ can be re-defined for histogram centers as follows
<rb>
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $ \textbf{x}=[x_{min} \ , \ x_{min}+\Delta{x} , \ x_{min}+ 2\Delta{x} \ , \ \dots \ ,  \ x_{max}]=[x_1 \ , \ x_2 \ , \ x_3 \ , \dots, \ x_i \ , \dots \ , \ x_{n_{bins}}],$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(3)</div>
</div>    
    
where $\Delta x$ is the distance between bin centers as follows
  
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $\Delta x=\dfrac{x_{max}-x_{min}}{n_{bins}}$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(4)</div>
</div>     
    
so
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $x_{i+1}=x_i+\Delta x$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(5)</div>
</div>   
    

where $i$ is the bin index of the histogram.

<br> Similarly, the probability mass function of $y$ can be then denoted as $p_y(y)$ where
    
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $y=[\ y_{min} , \  \ y_{min}+\Delta{y} \ , \  y_{min}+2\cdot \Delta{y} \ , \ \dots \ ,  \ y_{max}]=[y_1 \ , \ y_2 \ , \ y_3 \  , \dots \ , \ y_i \ , \dots, \  y_{n_{bins}}]$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(6)</div>
</div>  
and    
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $\textbf{p}_y=[{p_y(y) \ \forall \ y}] = [p_y(y_1) \ , \ p_y(y_2) \ ,\dots ,  \ p_y(y_i) \ , \dots \ ,\ p_y(y_{n_{bins}})],$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(7)</div>
</div>
    
where $\{ y_i \ ; \ i=1,2,...,n_{bins}\} $ are bin centers.     

<br>
<p style="text-align:justify;padding-right:5%">If individual data samples of two random variables $X$ and $Y$ have the same dimension, e.g. as for time series in (1) and (Fig.1), and if the numbers of bins are defined the same for $X$ and $Y$, then probability mass functions of $X$ and $Y$ are defined on the <b> same probability space </b> (here via the same 1-D bin indexing $i=1,2,...,n_{bins}$).</p>
<p style="text-align:justify;padding-right:5%">    
Similarly, the <b> same probability space </b> can be defined via the same numbers of bin centers of multidimensional probabilities (via multidim. histograms) for multi-dimensional random variables $X$ and $Y$ , i.e., where individual data samples are $n$-D points $\textbf{x}(k)=\left[ x_1(k) \ x_2(k) \ ... \ x_n(k) \right]$,$\textbf{y}(k)=\left[ y_1(k) \ y_2(k) \ ... \ y_n(k) \right]$, and  all observed data samples are $\textbf{X}=\left[ \textbf{x}(k=1) \ \textbf{x}(k=2), \ ..., \ \textbf{x}(k=N_x) \right] $,$\textbf{Y}=\left[ \textbf{y}(k=1) \ \textbf{y}(k=2), \ ..., \ \textbf{y}(k=N_y) \right] $, while the numbers of samples may be different for $X$ and $Y$, i.e., observed data are 2-D arrays (matrices) with dimensions $N_x\times n, \ N_y\times n $ where not necesarilly $N_x=N_y$. (just recall that for n-dimensional random variable $X$ and thus for n-dimensional probability $p_x(X=\textbf{x})$ it must hold that<p>
  <div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\sum\limits_{i_1=1}^{n_{bins,1}}\sum\limits_{i_2=1}^{n_{bins,2}}\dots\sum\limits_{i_n=1}^{n_{nbins,n}}p_\textbf{x}(i_1,i_2,\dots,i_n)=1$ ,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(8)</div>
</div> <br>
<p style="text-align:justify;padding-right:5%">where $i_j$ is bin center indexing for dimensions $j=1,2,...,n$  and for custom defined numbers of bins $n_{bins,j}$. If $n_{bins,j}$ are the same also for $n$-dimensional random variable $Y$, then $X$ and $Y$ have the <b>same probability space</b>.</p>
   
    
The following sketch and code with plots further explains the above notation.


    
<center><b>Fig. 2</b>: Histogram's $i$-th bin and the bin centre. (The bins define the probability space for $x$, basicaly by the number of bins $n_{bins}$.) </center>   
    
<div style="width:100%;height:1px;background-color:#408d62;margin-top:15px"></div>


```python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
rcParams.update({'font.size': 12})
#======
def sr(a):  # for plots
    a=str(round(a,2))
    return(a)
#======
Nx=500
Ny=1000
n_bins=20
#----------
x=random.randn(Nx)/3+1
y=random.binomial(Ny, .01, size=Ny)*2-10  # y=0,1,2,...10000
#y=(y-mean(y))/std(y)  #z-scoring, data normalization

freq_x,x_edges=histogram(x,n_bins)  #
freq_y,y_edges=histogram(y,n_bins)  #

dx=(max(x)-min(x))/n_bins   # Delta x
dy=(max(y)-min(y))/n_bins   # Delta x

x_centers=[x_edges[i]+dx/2 for i in range(n_bins)]
y_centers=[y_edges[i]+dy/2 for i in range(n_bins)]


px=freq_x/Nx  # probabilities
py=freq_y/Ny


figure(figsize=(14,10))
subplots_adjust(hspace=0.8)
subplot(421);plot(x,label="x(k)");grid();ylabel("x(k)");xlabel("k [sample]");ylim(-1,4);xlim(0,1000);legend()
title("(Measured) samples of $X$ in time (sample domain) \n $N_x=$"+str(Nx)+"$ \ , \ \overline{x}=$"+sr(mean(x))+"$ \ ,\  \sigma_x=$"+sr(std(x)))
subplot(422);plot(y,label="y(k)");grid();ylabel("y(k)");xlabel("k [sample]");ylim(-30,30);legend()
title("(Measured) samples of $Y$ in time (sample domain) \n $N_y=$"+str(Ny)+"$ \ , \ \overline{y}=$"+sr(mean(y))+"$ \ ,\  \sigma_y=$"+sr(std(y)))
subplot(423);hist(x,bins=n_bins);grid();xlabel("x");ylabel("frequencies");title("histogram of $X \ , \ n_{bins}=$"+str(n_bins));
subplot(424);hist(y,bins=n_bins);grid();ylabel("frequencies");xlabel("y");title("histogram of $Y \ , \ n_{bins}=$"+str(n_bins))
subplot(425);plot(x_centers,px,'-o');title("$p_x(x) \ , \ \sum p_x(x)=$"+str(sum(px)));grid();xlabel("$x$ [/]")
subplot(426);plot(y_centers,py,'-o');title("$p_y(y) \ , \ \sum p_y(y)=$"+str(sum(py)));grid();xlabel("$y$ [/]")
subplot(427);plot(arange(n_bins)+1,px,'-o');title("$p_x(i) \ , \ \sum p_x(i)=$"+str(sum(px))+" , $ n_{bins}=$"+str(n_bins));grid();xlabel("$i$ [bin index]")
subplot(428);plot(arange(n_bins)+1,py,'-o');title("$p_y(i) \ , \ \sum p_y(i)=$"+str(sum(py))+" , $ n_{bins}=$"+str(n_bins));grid();xlabel("$i$ [bin index]")
show()
```

<center><b>Fig. 3</b>: Probabilities for sets $x$ and $y$ at the same probability space, i.e., $n_{bins,x}=n_{bins,y}=n_{bins}$. </center>


<b>We are now to compare the distributions</b> of $X$ and $Y$, i.e., to compare their probability mass functions $p_x(i)$ and $p_y(i)$ that can tell us how similar or distinct is behavior of these data regardless the length of data, and regardless their means and magnitudes.

Notice that to have the same probability space for $X$ and $Y$, both $X$ and $Y$ have to be of same-dimensionality (here are both 1-D arrays) while their length can be different; however, the number of bins must be the same.

<h3> Relative Entropy (Kullback-Leibler divergence) </h3>


Let's have two random variables $X$ and $Y$ whose values are denoted $x(k)$ and $y(k)$ in sample domain, and $x_i$ and $y_i$. The shapes of their probability distributions can be roughly estimated from their histograms.

Let's denote their probability mass functions accordingly as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $p_x=p_x(x_i) \  \text{ or } \  p_y=p_y(y_i)$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(9)</div>
</div>

or when using only the bin indexing $i$ as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $p_x=p_x(i) \  \text{ or } \  p_y=p_y(i)$ </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(10)</div>
</div>

where $i=1,2,3,...,n_{bins}$.

Then, the information content of the random variable outcome (falling into $i$-th bin) is accordingly as follows
<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $I_x(x_i)=log_2\left(\dfrac{1}{p_x(x_i)}\right)\  , \ I_y(y_i)=log_2\left(\dfrac{1}{p_y(y_i)}\right) \ \text{[bit]}$ </div>
    <div style="width:10%;height:40px;float:left;text-align:right">(11)</div>
</div>


or analogically when using only $i$ for the bin indexing as

<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $I_x(i)=-log_2\big(p_x(i)\big) \ , \ I_y(i)=-log_2\big(p_y(i)\big) \ \text{[bit]} \text{ where } i=1,2,...,n_{bins}$ </div>
    <div style="width:10%;height:40px;float:left;text-align:right">(12)</div>
</div>






### Gibbs' inequality and KL Divergence



<center><b>Fig. 4</b>: Probabilities for sets $x$ and $y$ at the same probability space, i.e., $n_{bins,x}=n_{bins,y}=n_{bins}$. </center>

Because both $p_x,p_y \in \langle 0 ,1 \rangle$, it always holds for Shannon entropy that

<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\sum\limits_{i=1}^{i=n_{bins}}p_x(i)\cdot I_x(i)\le \sum\limits_{i=1}^{i=n_{bins}}p_x(i)\cdot I_y(i)\big)$, </div>
    <div style="width:10%;height:40px;float:left;text-align:right">(13)</div>
</div>
i.e.
<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $0\le\sum\limits_{i=1}^{i=n_{bins}}p_x(i)\cdot log_2\big(p_x(i)\big) - \sum\limits_{i=1}^{i=n_{bins}}p_x(i)\cdot log_2\big(p_y(i)\big)=\dots=\sum\limits_{i=1}^{i=n_{bins}}p_x(i)\cdot log_2\left(\dfrac{p_x(i)}{p_y(i)}\right)=D_{KL}(p_x||p_y) $ [bit] ,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(14)</div>
</div>

where $D_{KL}$ is the <b>relative entropy</b> that is also called the <b>Kullback-Leibler divergence</b>, and $D_{K L}=0$ only if the both distributions $p_x$ and $p_y$ are equal.

$D_{KL}$ is practically useful,eventhough the KL divergence is not a true distance between two distributions because it is not symetric, i.e.,  $D_{K L}(p_x||p_y)\ne D_{K L}(p_y||p_x)$  and does not respect the tringular inequality, e.g., p.18 in [2].

Also notice, that $p_x$ and $p_y$ can be two differrent distributions, obtained from two different data sets, but they have to be defined at the same probability space (here both 1-D arrays and $i=1,2,...,n_{bins,x}=n_{bins,y}$).

For KL divergence, study also section 2.6 in [1] and related sections.

```python
"""1) Generate 300 samples of 1-D random variable  X , i.e.,  X=x(k)  where  k=1,2,...,Nx=300  with normally distributed data with mean  x=50
    and standard deviation  x=10  , draw the data in a sample domain and draw the data in the probability domain (probability mass function  px(x))"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mean = 50
std_dev = 10
Nx = 300

# Generate samples
samples_X = np.random.normal(mean, std_dev, Nx)

# data in the sample domain
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(range(1, Nx + 1), samples_X, color='b', alpha=0.6)
plt.axhline(mean, color='r', linestyle='--', label=f'Mean = {mean}')
plt.xlabel('Sample Index')
plt.ylabel('Sample Value')
plt.title('Data in Sample Domain')
plt.legend()

# data in the probability domain (approximating the PMF with a histogram)
plt.subplot(1, 2, 2)
count, bins, ignored = plt.hist(samples_X, bins=15, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.plot(bins, norm.pdf(bins, mean, std_dev), 'r--', linewidth=2, label='Normal Distribution Fit')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Data in Probability Domain (PMF approximation)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
"""2) Generate 500 samples of 1-D random variable  Y , i.e.,  Y=y(k)  where  k=1,2,...,Ny=500  ( Ny ...number of generated samples) with binomial distribution
    with probability parameter  p=0.45  as in [a], number of trials  n=100 , e.g. use [a] , draw the data in a sample domain and draw the data in the probability domain
    (probability mass function  py(y) . Notice, the parameter  p  as in [a] is not the same as  py(y)  here"""

# Parameters
p = 0.45
n = 100
Ny = 500  # Number of samples

# Generate samples
samples_Y = np.random.binomial(n, p, Ny)

# data in the sample domain
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(range(1, Ny + 1), samples_Y, color='g', alpha=0.6)
plt.axhline(np.mean(samples_Y), color='r', linestyle='--', label=f'Mean = {np.mean(samples_Y):.2f}')
plt.xlabel('Sample Index')
plt.ylabel('Sample Value')
plt.title('Data in Sample Domain')
plt.legend()

# data in the probability domain (approximating the PMF with a histogram)
plt.subplot(1, 2, 2)
count, bins, ignored = plt.hist(samples_Y, bins=20, density=True, color='lightgreen', edgecolor='black', alpha=0.7)
plt.plot(bins, binom.pmf(bins.astype(int), n, p), 'r--', linewidth=2, label='Binomial PMF')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Data in Probability Domain (PMF approximation)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
"""3)For the above 1-D random variables  X  and  Y  calculate the K-L divergence  DKL(X||Y)"""

# range and bins for PMF estimation
xmin = min(min(samples_X), min(samples_Y))
xmax = max(max(samples_X), max(samples_Y))
bins = np.arange(xmin, xmax + 1)  # Use integer bins for clarity

# Calculate empirical PMFs for X and Y
hist_X, _ = np.histogram(samples_X, bins=bins, density=True)
hist_Y, _ = np.histogram(samples_Y, bins=bins, density=True)

# Avoid division by zero by adding a small epsilon to histograms
epsilon = 1e-10
hist_X = hist_X + epsilon
hist_Y = hist_Y + epsilon

# Calculate KL divergence
kl_divergence = entropy(hist_X, hist_Y)

print("KL Divergence D_KL(X || Y):", kl_divergence)
```

```python
# 4) For the above 1-D random variables  Y  and  X  calculate the K-L divergence  DKL(Y||X)

kl_divergence = entropy(hist_Y, hist_X)

print("KL Divergence D_KL(Y || X):", kl_divergence)
```

```python
""" 5) Keep the normally distributed random variable  X  from the above.
      Calculate and draw graph of  DKL(X,Y(p))  where the binomial distribution parameter  p  for random variable  Y  varies as  p=[0.1,0.2,0.3,...,0.9] """

p_values = np.arange(0.1, 1.0, 0.1)
kl_divergences = []

# Calculate KL divergence D_KL(X || Y(p)) for varying p
n_Y = 100  # Number of trials for Y
for p in p_values:
    samples_Y = np.random.binomial(n_Y, p, 500)  # Generate 500 samples for each p

    # Calculate empirical PMFs for X and Y
    hist_X, bins_X = np.histogram(samples_X, bins=30, density=True)
    hist_Y, _ = np.histogram(samples_Y, bins=bins_X, density=True)

    epsilon = 1e-10
    hist_X += epsilon
    hist_Y += epsilon

    # Calculate KL divergence
    kl_divergence = entropy(hist_X, hist_Y)
    kl_divergences.append(kl_divergence)

# Plotting the KL divergence against p
plt.figure(figsize=(10, 6))
plt.plot(p_values, kl_divergences, marker='o', linestyle='-', color='b')
plt.title('KL Divergence D_{KL}(X || Y(p)) for Varying p')
plt.xlabel('Probability Parameter p for Y')
plt.ylabel('KL Divergence D_{KL}(X || Y(p))')
plt.xticks(p_values)
plt.grid()
plt.show()
```

```python
""" 6) Instead of  DKL(X||Y(p))  as in the previous task, calculate and draw the Euclidean distance between the two distributions of p_x(X) and p_y(Y)
    for the generated data from the previous task, i.e. between observations normally distributed  X  and  Y  observations with binomial distribution of  Y
    for  p=[0.1,0.2,0.3,...,0.9] """



import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom

# Parameters for the normal distribution X
mean_X, std_dev_X = 50, 10
Nx = 300  # Number of samples for X

# Generate samples for X
samples_X = np.random.normal(mean_X, std_dev_X, Nx)

# Euclidean distance values for different p
p_values = np.arange(0.1, 1.0, 0.1)
euclidean_distances = []

# Calculate Euclidean distance for varying p
n_Y = 100  # Number of trials for Y
for p in p_values:
    # Generate 500 samples for the binomial distribution Y
    samples_Y = np.random.binomial(n_Y, p, 500)

    # Calculate empirical PMFs for X and Y
    hist_X, bins_X = np.histogram(samples_X, bins=30, density=True)
    hist_Y, bins_Y = np.histogram(samples_Y, bins=bins_X, density=True)

    # Calculate bin centers
    bin_centers_X = 0.5 * (bins_X[:-1] + bins_X[1:])
    bin_centers_Y = 0.5 * (bins_Y[:-1] + bins_Y[1:])

    # Ensure the histograms are the same length for comparison
    if len(hist_X) > len(hist_Y):
        hist_Y = np.pad(hist_Y, (0, len(hist_X) - len(hist_Y)), 'constant')
    elif len(hist_X) < len(hist_Y):
        hist_X = np.pad(hist_X, (0, len(hist_Y) - len(hist_X)), 'constant')

    #  Compute the Euclidean distance based on bin centers
    # Interpolate Y's histogram to the X's bin centers
    hist_Y_interp = np.interp(bin_centers_X, bin_centers_Y, hist_Y, left=0, right=0)

    # Calculate the Euclidean distance
    euclidean_distance = np.sqrt(np.sum((hist_X - hist_Y_interp) ** 2))
    euclidean_distances.append(euclidean_distance)

# Plotting the Euclidean distance against p
plt.figure(figsize=(10, 6))
plt.plot(p_values, euclidean_distances, marker='o', linestyle='-', color='b')
plt.title('Euclidean Distance between p_X(X) and p_Y(Y(p)) at Bin Centers')
plt.xlabel('Probability Parameter p for Y')
plt.ylabel('Euclidean Distance D_E(p_X, p_Y)')
plt.xticks(p_values)
plt.grid()
plt.show()

```

```python
""" 7) Here, investigate  DKL  vs. the Euclidean distance as measures for similarity of two distributions. Repeat the above simulation experiments,
  i.e. task 5. and 6., at least 10-times, and observe (plot curves of) the behavior of D_{KL} vs. the Euclidean distances behavior
  (for at least 10 various generated datasets). What is your findings when comparing these two measures?"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, entropy

# Parameters for the normal distribution X
mean_X, std_dev_X = 50, 10
Nx = 300  # Number of samples for X
n_Y = 100  # Number of trials for Y
num_experiments = 10  # Number of datasets to generate

# Probability parameter values for Y
p_values = np.arange(0.1, 1.0, 0.1)

# To store results
kl_results = []
euclidean_results = []

# Perform the experiments
for _ in range(num_experiments):
    samples_X = np.random.normal(mean_X, std_dev_X, Nx)  # Samples for X
    kl_distances = []
    euclidean_distances = []

    for p in p_values:
        samples_Y = np.random.binomial(n_Y, p, 500)  # Generate samples for Y

        # Calculate empirical PMFs for X and Y
        hist_X, bins_X = np.histogram(samples_X, bins=30, density=True)
        hist_Y, bins_Y = np.histogram(samples_Y, bins=bins_X, density=True)

        # Calculate bin centers
        bin_centers_X = 0.5 * (bins_X[:-1] + bins_X[1:])
        bin_centers_Y = 0.5 * (bins_Y[:-1] + bins_Y[1:])

        # Interpolate Y's histogram to the X's bin centers
        hist_Y_interp = np.interp(bin_centers_X, bin_centers_Y, hist_Y, left=0, right=0)

        # Avoid division by zero for KL divergence calculation
        epsilon = 1e-10
        hist_X += epsilon
        hist_Y_interp += epsilon

        # Calculate KL divergence
        kl_distance = entropy(hist_X, hist_Y_interp)
        kl_distances.append(kl_distance)

        # Calculate the Euclidean distance
        euclidean_distance = np.sqrt(np.sum((hist_X - hist_Y_interp) ** 2))
        euclidean_distances.append(euclidean_distance)

    kl_results.append(kl_distances)
    euclidean_results.append(euclidean_distances)

# Convert results to arrays for easier plotting
kl_results = np.array(kl_results)
euclidean_results = np.array(euclidean_results)

# Plotting
plt.figure(figsize=(14, 8))

# Plot KL divergence results
for i in range(num_experiments):
    plt.plot(p_values, kl_results[i], marker='o', linestyle='-', alpha=0.5)

plt.title('KL Divergence D_{KL}(X || Y(p)) for Varying p Across Experiments')
plt.xlabel('Probability Parameter p for Y')
plt.ylabel('KL Divergence')
plt.grid()
plt.legend(loc='upper left', fontsize='small')
plt.show()

plt.figure(figsize=(14, 8))

# Plot Euclidean distance results
for i in range(num_experiments):
    plt.plot(p_values, euclidean_results[i], marker='o', linestyle='-', alpha=0.5, color='orange', label=f'D_E Experiment {i+1}')

plt.title('Euclidean Distance D_E(p_X, p_Y) for Varying p Across Experiments')
plt.xlabel('Probability Parameter p for Y')
plt.ylabel('Euclidean Distance')
plt.grid()
plt.legend(loc='upper left', fontsize='small')
plt.show()
```

Observe the KL divergence and Euclidean distance vary with different values of p.

KL Divergence is sensitive to the tails of the distributions. It may increase sharply when p leads to a significant change in the distribution shape of Y(p).

Euclidean Distance can be affected by the distribution of mass across the bins. It might not capture the difference in distribution tails as effectively as KL divergence.
