 **Exercise 3**. You are given 16 balls, all of which are equal in weight except for one that is either heavier or lighter.
You are also given a bizarre two-pan balance that can report only two outcomes: ‘the two sides balance’ or ‘the two
sides do not balance’. Design a strategy to determine which is the odd ball in as few uses of the balance as possible.

**Solution**: At each weighing, the balance provides two possible outcomes: balanced or not balanced. Since there are two outcomes per weighing, the number of scenarios we can differentiate after n weighings is 3^n. This accounts for identifying the odd ball and determining whether it is heavier or lighter, as 3^n includes both heavier and lighter possibilities.
For 16 balls, we need n weighings such that 3^n ≥ 2 x 16=32 (to cover all possibilities for the odd ball being either heavier or lighter). Thus, n=4 weighings are sufficient.

**Steps for the solution:**

First Weighing:
Divide the 16 balls into three groups: 5, 5, and 6.
Weigh the first group (5 balls) against the second group (5 balls).
If balanced, the odd ball is in the unweighed group (6 balls).
If unbalanced, the odd ball is in the heavier or lighter group.

Second Weighing:
Take the group identified in the first weighing and divide it into three smaller groups as evenly as possible.
Example: If there are 6 balls, divide into groups of 2, 2, and 2.
Example: If there are 5 balls, divide into groups of 2, 2, and 1.

Weigh two of these groups.
If balanced, the odd ball is in the unweighed group.
If unbalanced, the odd ball is in the heavier or lighter group.

Third Weighing:
Repeat the process of dividing the identified group into three smaller groups and weighing two of them.

Fourth Weighing:
By now, we will have isolated the odd ball to a single candidate. Perform the final weighing to determine if it is heavier or lighter.

 **Exercise 6**. Let PY = {0.5, 0.5}. Sketch 1/N H δ(Y N ) as a function of δ for N = 1, 2, 3 and 100.

```python
import numpy as np
import matplotlib.pyplot as plt

PY = [0.5, 0.5]  # Probabilities for Y
H_Y = -sum(p * np.log2(p) for p in PY)  # Shannon entropy of Y
delta_values = np.linspace(0, 2, 500)  # Range of delta values
N_values = [1, 2, 3, 100]

# Calculate 1/N * H_delta(Y^N) as a function of delta for each N
results = {}
for N in N_values:
    H_delta_YN = delta_values * H_Y  # Entropy scaled by delta
    results[N] = H_delta_YN / N      # Scaled entropy divided by N

plt.figure(figsize=(10, 6))
for N in N_values:
    plt.plot(delta_values, results[N], label=f"N = {N}")

plt.title("Scaled Entropy 1/N * H_delta(Y^N) vs Delta", fontsize=14)
plt.xlabel("Delta (δ)", fontsize=12)
plt.ylabel("1/N * H_delta(Y^N)", fontsize=12)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
