### Some "Refreshment"

Let's recall conditional probability via the flying drone example, where the drone can physically fly only inside the area $A$ and $B$ and these two areas are overlapping.

The probability that the drone randomly crashes are as follows:
- Marginal probability is probability:<br>
  $P(A)$ ... probability that the drone crashed in area $A$<br>
  $P(B)$ ... probability that the drone crashed in area $B$
- Joint probability is probability:<br>
  $P(A,B)=P(B,A)=P(AB)=P(BA)$ ... probability that the drone crashed both in $A$ and $B$ , i.e. in the intersection of the two areas (if areas do overlap) <br>
- Conditional probability is probability:<br>
  $P(A|B)$ ... probability that the drone crashed in area $A$ while it is given it was flying only in area $B$ (because the areas overlap)
  $P(B|A)$ ... probability that the drone crashed in area $B$ while it is given it was flying only in area $A$ (because the areas overlap)

Notice, that for particual values of probability and for probability mass function (of a discrete random variable) a small $p$ is usually used. Capital $P$ is usually used to denote probability in general (such as for theoretical formulas etc).


<center><b>Fig. 1</b>: Flying drone example and its event space to recall the conditional probability concept. </center>

Also, recall here the Bayesian rule as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $P(A|B)\cdot P(B)=P(B|A)\cdot P(A)$ ,</div>
    <div style="width:10%;height:20px;float:left;text-align:right">(1)</div>
</div>

which can be demonstrated on the drone example above.

And further, let's recall the general formula for complete probability as follows

<div style="width:100%;height:30px;margin-top:10px">
    <div style="width:90%;height:30px;float:left;text-align:center">
    $P(B)=\sum\limits_{i=1}^{i=N_A} P(B|A_1)\cdot P(A_i)$ ,</div>
    <div style="width:10%;height:30px;float:left;text-align:right;padding:10px 0 0 0;">(2)</div>
</div> <br>

which can be demonstrated, e.g., on tossing two dices, where $A_i$ can be outcome that dice A turns number "$i$" where $i=1,2,3,..6$, and $B$ can represent the outcome that dice B turns "6".  

The principles in formulas (1) and (2) can be then combined for complete Bayes rule(s).

<h2> Data Dependency </h2>

<h3>Multidimensional, multivariate data dependency </h3>

Let's generally define a function $f()$ of multidimensional feature vector as follows
    
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $y(k)=f\big(\textbf{x}(k)\big)$, </div>
    <div style="width:10%;height:20px;float:left;text-align:right">(3)</div>
</div>

 where the input term $\textbf{x}(k)$ is the feature vector, the output $y(k)$ remains scalar here, $k$ is sample index or discrete time index, and $\textbf{X}$ would be then the matrix of all features as in Fig.2.

In this case, $\textbf{x}(k)$ may represent a feature vector that can serve as an input to a neural network, and $y(k)$ may represent the corresponding label (target, or neural output, ...).



<center style="text-align:justify"><b>Fig. 2</b>: See the file "Xy.csv" in subfolder "data" of this Jupyter Notebook. Here, $\textbf{X}$ denotes a matrix with row feature vectors , so we look at the 9-dimensional feature vector as at a 9-dimensional random variable $\textbf{x}$  (Recall: $\textbf{X},\textbf{x},x$...matrix, vector, scalar)</center>


For a multivariate mapping (3) and data example "Xy.csv" in Fig.2, there is a 9-dimensional feature vector $\textbf{x}(k)$ and a scalar label $y(k)$, then:

- the probability of $k$-th feature vector instance in sample-index domain is $p_\textbf{x}\big(\textbf{x}(k)\big)=p_\textbf{x}\big(x_1(k),x_2(k),\dots,x_9(k)\big)$, and for a histogram it is as $p_\textbf{x}(i_1,i_2,\dots,i_9)$, so the probability of $\textbf{x}$, is a hypersurface in $(9+1)$-dimensional space (for $\textbf{x}$ being 9-element long vector), for which

<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\sum\limits_{i_1=1}^{n_{bins,1}}\sum\limits_{i_2=1}^{n_{bins,2}}\dots\sum\limits_{i_9=1}^{n_{nbins,9}}p_\textbf{x}(i_1,i_2,\dots,i_9)=1$ ,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(4)</div>
</div> <br>

 $ \ \ \ \  \ \ \ $ where the number of bins $n_{bins,j}$ for features $x_j$ where $j=1,2,...,9 $ can be estimated for each feature $x_1,x_2,...,x_j,...x_9$, individually.


- the probability of  $k$-th label in sample-index domain is $p_y\big(y(k)\big)$, and for a histogram it is as $p_y(i) \text{ where } i=1,2,...,n_{bins,y}$, so the probability of $y$ is a curve in 2-D.

Now we see that the feature vector $\textbf{x}$ and the scalar label $y$ <b>do not have the same probability space, so we can not use $D_{KL}$ </b> to evaluate the similarity of their distributions, i.e., we can not evaluate the relationships (3) via distribution comparison of $\textbf{x}$ and  $y$ by $D_{KL}$. <br>
(Obviously , we can not use Pearson's (linear) correlation coefficient $r$ either, because $r$ is defined on for scalar variables of the same length ("same sample domain").)

### Some remarks: What we can do to find relationship of multidimensional data data?

- Basically, we may try to (batch) train some multivariate regressor, such as linear or polynomial neuron or some other type of shallow neural architecture, preferably linear in parameters, and test its performance, that can give us the insight how well are the features correlated to the labels and if our configuration of training data is ok.

- Obviously, this can be suitable for problems <b> where deep networks and deep learning can not be done</b> or does not make sense, e.g., if we do not have enough of training data, or if we need to apply machine learning tasks where computation can not be done with sufficiently powerful HW.

- Notice, the more powerfull neural architecture you use, the better you fit the traning data, however, it does not necessarilly mean, that such trained neural architecture grasped the dependency in data correctly, i.e., <b>be aware of overfitting (overtraining) vs. good generalization of the neural networks. </b>

- Notice, that we may use some feature selection method (e.g., BORUTA [$\alpha$]) that can help us to identify the best features,  but then we still might to wish to double check the feature design, e.g., to validate whether using only features e.g., $x_6,x_8$ is better than combination of other two features to predict label $y$.

- We may use <b>Mutual Information</b> that is suitable for evaluation of nonlinear multidimensional relationships (3) on different probablity spaces.

<h3> Mutual Information (MI)</h3>

<b>Mutual Information (MI)</b> is suitable for evaluating the strength of </b> nonlinear multidimensional (multivariate) </b> relationships of data on different probablity spaces.

The input (features) are vectors and the output (label, target) can be either scalar, such as in (3), Fig. 2,  or vectors too, such as a neural network with multiple output neurons.

Recall mapping (3) is as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $y(k)=f\big(\textbf{x}(k)\big)$, </div>
    <div style="width:10%;height:20px;float:left;text-align:right"></div>
</div>

and the mapping would be for a fully multivariate problem
<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $\textbf{y}(k)=\textbf{f}\big(\textbf{x}(k)\big)$,</div>
    <div style="width:10%;height:20px;float:left;text-align:right">(5)</div>
</div>

where  $f()$ would yield a vector function $\textbf{f}(\ )$ as follows

<div style="width:100%;height:90px;margin-top:10px">
    <div style="width:90%;height:90px;float:left;text-align:center">
      $ \textbf{y}(k)=\begin{bmatrix} y_1(k) \\ y_2(k) \\ \vdots \\ y_{ny}(k) \end{bmatrix} \ , \
      \textbf{f}(k)=\begin{bmatrix} f_1(\textbf{x}(k) ) \\ f_2(\textbf{x}(k)) \\ \vdots \\ f_{ny}(\textbf{x}(k)) \end{bmatrix} \ , \
      \textbf{x}(k)=\begin{bmatrix} x_1(k) \\ x_2(k) \\ \vdots \\ x_{nx}(k) \end{bmatrix} \
       ,$
    </div>
    <div style="width:10%;height:90px;float:left;text-align:right;padding:40px 0 0 0">(6)</div>
</div>

where, for example, we may generally think of $\textbf{f}(\ )$ as of a neural network with $n_y$ output neurons.

Further for clarity, let's continue with scalar $Y$, i.e. $y$ as in (3), i.e., for the example of data relationship as in Fig.2.

Recall that probabilities we may calculate for data as for (3), Fig.2, are: <br>
- marginal probabilities $p(\textbf{x})$, $p(y)$,
- joint probabilities $p(\textbf{x},y)=p(y,\textbf{x})$, and<br>
- conditional probabilities $p(y|\textbf{x}) $, $p(\textbf{x}|y) $.

 <br>
-  information marginal content $h(X)=I(X),h(Y)=I(Y)$,
-  information joint content  $h(X,Y)=I(X,Y)$ (notice the comma "," separator for joint events, recall that
  - $P(X,Y)=P(XY)=P(X \ and \ Y)$ and that  $P(XY)=P(YX)$  
- information conditional content $h(X|Y)=I(X|Y),h(Y|X)=I(Y|X)$ and

 <br>
- marginal entropies $H(X),H(Y)$,
- joint entropy $H(X,Y)$,
- conditional entropies $H(X|Y),H(Y|X)$ and
- <b style="color:blue">Mutual Information </b>$I(X\color{blue}{;}Y)$ (notice the semicolon "<b>;</b>" as separator for MI!!, "H" is not used for "MI")

and once recalling the formula for information content $h(x)=I(x)=log\left(\dfrac{1}{p(x)} \right)=-log\big(p(x)\big)$ and for the (marginal) entropy as its average via probability mass function as $H(x)=\sum\limits_{\forall x} p(x) h(x)$  , the extensions to joint and conditional entropies should be obvious (find them in [1]).

<b> Important:</b> For MI, study section 8.1 in [1] and related parts, see also Fig. 8.1!

The brief recall on derivation of MI can start with the fundamental Bayes rule for two random variables as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $P(Y,X)=P(Y|X)\cdot P(X)=P(X|Y)\cdot P(Y)$,</div>
    <div style="width:10%;height:20px;float:left;text-align:right"><p style="color:blue;text-align:right">(7)</p></div>
</div>

where by applying a minus sign and logarithm to (7), i.e., $-log\big(\color{blue}{\text{(7)}}\big)$, we arrive to the <b>chain rule of information content </b> $h$ as follows


<div style="width:100%;height:90px;margin-top:10px">
    <div style="padding-left:10%;width:90%;height:90px;float:left;text-align:justify">
    $-log\big(P(Y,X)\big)=-log\big(P(Y|X)\big)-log\big(P(X)\big)=-log\big(P(X|Y)\big)-log\big(P(Y)\big)=$ <br>
        $=log\left(\dfrac{1}{P(X,Y)}\right)=log\left(\dfrac{1}{P(Y|X)}\right)+log\left(\dfrac{1}{P(X)}\right)=log\left(\dfrac{1}{P(X|Y)}\right)+log\left(\dfrac{1}{P(Y)}\right)=$<br>
        $=h(Y,X)=h(Y|X)+h(X)=h(X|Y)+h(Y)$
        </div>
    <div style="width:10%;height:90px;float:left;text-align:right;padding:40px 0 0 0">(8)</div>
</div>

where $h()$ stand for the Shannon information content.

The chain rule of $h()$ in (8) leads us to the <b> chain rule of entropy </b> as follows

<div style="width:100%;height:20px;margin-top:10px">
    <div style="width:90%;height:20px;float:left;text-align:center">
    $H(Y,X) = H(Y|X)+H(X)  = H(X|Y)+H(Y)$,</div>
    <div style="width:10%;height:20px;float:left;text-align:right">(9)</div>
</div>

and if we substract both conditional entropies $H(X|Y)$ and $H(Y|X)$ from every part of (9), we obtain theoretical formula for the Mutual Information as follows


<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $I(Y;X)=I(X;Y)=H(X,Y) − H(X | Y )- H(Y | X )=$ <br>$= H(X) − H(X | Y )=H(Y) − H(Y | X )$,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(10)</div>
</div>

i.e., Mutual Information is symetric $I(X; Y ) = I(Y ; X)$ and non-negative $I(X; Y ) ≥ 0$, see Fig.8.1 in [1].

See (8.3) on p. 139 in [1] for ilegality of three-term MI.

### MI in Probability Space
Notice, MI can be calculated when $X$,$Y$ also have different probability spaces.

First,  one-dimensional, ${n_x}$-dimensional (if random variable $X$ is a vector $\textbf{x}$ of length $n_x$, e.g., $n_x=9$ in Fig.2), and $(n_x+1)$-dimensional probability mass functions are calculated from measured data (in the sample space), either via multidimensional histograms, or more often by clustering, or by some kernel technique estimation [b] (some rough estimations can be done via a simple RBF network too ... it can be a project for this class)

Then, the formula for computing the Mutual Information in the probability space is as follows

<div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $I(X;Y)=I(Y;X)=\sum\limits_{\forall\textbf{x} \in X}\sum\limits_{\forall y \in Y }p_{\textbf{x},y}(\textbf{x},y)\cdot log\left(\dfrac{p_{\textbf{x},y}(\textbf{x},y)}{p_\textbf{x}(\textbf{x})\cdot p_y(y)} \right)$,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(11)</div>
</div>

where, either multi-dimensional histograms and probabilities have to be calculated so that

  <div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\sum\limits_{i_1=1}^{n_{bins,1}}\sum\limits_{i_2=1}^{n_{bins,2}}\dots\sum\limits_{i_n=1}^{n_{nbins,n}}p_\textbf{x}(i_1,i_2,\dots,i_n)=1$ ,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(12)</div>
</div> <br>

  <div style="width:100%;height:40px;margin-top:10px">
    <div style="width:90%;height:40px;float:left;text-align:center">
    $\sum\limits_{i_1=1}^{n_{bins,1}}\sum\limits_{i_2=1}^{n_{bins,2}}\dots\sum\limits_{i_n=1}^{n_{nbins,n}}\sum\limits_{j=1}^{n_{nbins,y}}p_{\textbf{x},y}(i_1,i_2,\dots,i_n,j)=1$,</div>
    <div style="width:10%;height:40px;float:left;text-align:right">(13)</div>
</div> <br>



### Practical Estimation of Nonlinear Relationship for Large Datasets (MI formula via clustering)

Practically, we may estimate the probability densities ([b] roughly by a RBF layer) , and estimate the MI for our data,  or more practically using existing librarties, MI can be estimated via clustering techniques that might be faster to calculate than multidimensional histograms, e.g., see  [$\beta$].

$$
MI(U,V)=\sum_{i=1}^{|U|} \sum_{j=1}^{|V|} \frac{|U_i\cap V_j|}{N}\log\frac{N|U_i \cap V_j|}{|U_i||V_j|}
$$

where $|U_i|$ is the number of the samples in cluster $U_i$ and $|V_i|$ is the number of the samples in cluster $V_i$ (see [$\beta$]

```python
"""1)Evaluate the strength of linear relationship between observed values of variables  Y  and  X  in data file data\yx_linear.txt
    and in data file data\yx_nonlinear.txt.In particular, implement in your own code (via Numpy) the Pearson's correlation coefficient
    Validate your result with some existing tool, e.g. numpy.corrcoef().Plot function  y=y(x)  for both data sets.
    Very briefly compare/explain your results in terms of linearity vs. nonlinearity of data.
"""
import numpy as np
import matplotlib.pyplot as plt

data_linear = np.loadtxt('/content/xy_linear.txt', skiprows=1)  # Skipping header row
data_nonlinear = np.loadtxt('/content/yx_nonlinear.txt', skiprows=1)

# Separate x and y values for both datasets
x_linear, y_linear = data_linear[:, 0], data_linear[:, 1]
x_nonlinear, y_nonlinear = data_nonlinear[:, 0], data_nonlinear[:, 1]

# Calculate Pearson correlation for linear dataset
N_linear = len(x_linear)
mean_x_linear, mean_y_linear = np.mean(x_linear), np.mean(y_linear)
std_x_linear, std_y_linear = np.std(x_linear), np.std(y_linear)
pearson_r_linear = np.sum((x_linear - mean_x_linear) * (y_linear - mean_y_linear)) / (N_linear * std_x_linear * std_y_linear)

# Calculate Pearson correlation for nonlinear dataset
N_nonlinear = len(x_nonlinear)
mean_x_nonlinear, mean_y_nonlinear = np.mean(x_nonlinear), np.mean(y_nonlinear)
std_x_nonlinear, std_y_nonlinear = np.std(x_nonlinear), np.std(y_nonlinear)
pearson_r_nonlinear = np.sum((x_nonlinear - mean_x_nonlinear) * (y_nonlinear - mean_y_nonlinear)) / (N_nonlinear * std_x_nonlinear * std_y_nonlinear)

# Validate with numpy's corrcoef
pearson_r_linear_np = np.corrcoef(x_linear, y_linear)[0, 1]
pearson_r_nonlinear_np = np.corrcoef(x_nonlinear, y_nonlinear)[0, 1]

print("Manual Pearson's r (Linear):", pearson_r_linear)
print("Numpy Pearson's r (Linear):", pearson_r_linear_np)
print("Manual Pearson's r (Nonlinear):", pearson_r_nonlinear)
print("Numpy Pearson's r (Nonlinear):", pearson_r_nonlinear_np)

# Plot y vs x for both datasets
plt.figure(figsize=(14, 6))

# Linear dataset plot
plt.subplot(1, 2, 1)
plt.scatter(x_linear, y_linear, color='blue', alpha=0.6)
plt.title("Linear Dataset: y vs. x")
plt.xlabel("x")
plt.ylabel("y")

# Nonlinear dataset plot
plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='red', alpha=0.6)
plt.title("Nonlinear Dataset: y vs. x")
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

For the linear dataset, we expect a high absolute value for r, close to 1 or -1, indicating a strong linear relationship. The plot of Y against X for the linear dataset should exhibit a clear linear pattern.

```python
""" 2) a) Evaluate the strength of nonlinear relationships between observed values of scalar variables  Y  and  X
      in data file data\yx_linear.txt and in data file data\yx_nonlinear.txt Mutual Information:
      Calculate MI with some existing library that directly returns MI and uses classical approach with probabilities (reference it)"""

from sklearn.feature_selection import mutual_info_regression
import numpy as np

# Separate x and y values for both datasets
x_linear1, y_linear1 = data_linear[:, 0].reshape(-1, 1), data_linear[:, 1]
x_nonlinear1, y_nonlinear1 = data_nonlinear[:, 0].reshape(-1, 1), data_nonlinear[:, 1]

mi_linear = mutual_info_regression(x_linear1, y_linear1)[0]
mi_nonlinear = mutual_info_regression(x_nonlinear1, y_nonlinear1)[0]

print("Mutual Information (Linear dataset):", mi_linear)
print("Mutual Information (Nonlinear dataset):", mi_nonlinear)

""" b) Calculate MI with your own code in probability space as by formula (11) for both datasets,
      you can use 1-D and n-D histogram functions from relevant modules to get probabilities
"""
import numpy as np

def mutual_information(x, y, bins=30):
    # Calculate joint histogram
    joint_hist, x_edges, y_edges = np.histogram2d(x, y, bins=bins)
    joint_prob = joint_hist / np.sum(joint_hist)  # Normalize to get joint probability distribution

    # Calculate marginal histograms and probabilities
    x_marginal = np.histogram(x, bins=x_edges)[0] / len(x)  # P(X)
    y_marginal = np.histogram(y, bins=y_edges)[0] / len(y)  # P(Y)

    # Broadcast the marginal distributions to match the joint distribution shape
    x_marginal = x_marginal.reshape(-1, 1)  # reshape for broadcasting
    y_marginal = y_marginal.reshape(1, -1)

    # Calculate Mutual Information
    non_zero_joint = joint_prob > 0  # Avoid log(0) by excluding zero probabilities
    mi = np.sum(joint_prob[non_zero_joint] *
                np.log(joint_prob[non_zero_joint] / (x_marginal * y_marginal)[non_zero_joint]))

    return mi

x_linear2, y_linear2 = data_linear[:, 0], data_linear[:, 1]
x_nonlinear2, y_nonlinear2 = data_nonlinear[:, 0], data_nonlinear[:, 1]

mi_linear1 = mutual_information(x_linear2, y_linear2, bins=30)
mi_nonlinear1 = mutual_information(x_nonlinear2, y_nonlinear2, bins=30)

print("\nMutual Information (Linear dataset) with own code:", mi_linear1)
print("Mutual Information (Nonlinear dataset) with own code:", mi_nonlinear1)

#c) Calculate the MI with some clustering-based method (with some existing library, e.g. [ β ] )

from sklearn.feature_selection import mutual_info_regression
import numpy as np

# Separate x and y values for both datasets
x_linear3, y_linear3 = data_linear[:, 0].reshape(-1, 1), data_linear[:, 1]
x_nonlinear3, y_nonlinear3 = data_nonlinear[:, 0].reshape(-1, 1), data_nonlinear[:, 1]

# Calculate Mutual Information using sklearn's mutual_info_regression with k-NN approach
mi_linear_knn = mutual_info_regression(x_linear3, y_linear3, n_neighbors=5)[0]
mi_nonlinear_knn = mutual_info_regression(x_nonlinear3, y_nonlinear3, n_neighbors=5)[0]

# Display the results
print("\nMutual Information (Linear dataset, k-NN based):", mi_linear_knn)
print("Mutual Information (Nonlinear dataset, k-NN based):", mi_nonlinear_knn)

```

```python
"""3) a) Based on Fig.2 and for dataset data\Xy.txt Calculate Mutual Information  I(X;Y)=I(x;y)  for a feature vector defined as
      x=[x6 x8]  and for the corresponding label  y : using some existing tool for MI"""

import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_selection import mutual_info_classif
import numpy as np

data = np.loadtxt('/content/XY.txt', skiprows=1)

# Extract columns for features x6, x8, and label y (last column)
x6 = data[:, 5]
x8 = data[:, 7]
y = data[:, -1]

# Stack x6 and x8 to form a feature matrix
X = np.column_stack((x6, x8))

mi_value = mutual_info_classif(X, y, discrete_features=True)[0]

print("Mutual Information I(X; Y) for x=[x6, x8] and label y:", mi_value)


# b)with your own code for MI in probability space as by formula (11)

def mutual_information_probability_space(x, y, bins=10):
    # Create 3D histogram for joint distribution P(x6, x8, y)
    hist_3d, edges = np.histogramdd(np.column_stack((x, y)), bins=(bins, bins, bins))
    joint_prob = hist_3d / np.sum(hist_3d)  # Normalize to get joint probabilities

    # Marginal distributions
    marginal_x = np.sum(joint_prob, axis=2)
    marginal_y = np.sum(joint_prob, axis=(0, 1))

    # Initialize MI
    mi = 0.0

    # Iterate over all non-zero joint probabilities
    for i in range(joint_prob.shape[0]):
        for j in range(joint_prob.shape[1]):
            for k in range(joint_prob.shape[2]):
                # Joint probability for specific (x6, x8, y) bin
                p_xy = joint_prob[i, j, k]
                if p_xy > 0:  # Avoid log(0)
                    # Marginal probabilities
                    p_x = marginal_x[i, j]
                    p_y = marginal_y[k]
                    # Mutual Information increment
                    mi += p_xy * np.log(p_xy / (p_x * p_y))

    return mi

x_6 = data[:, 5]
x_8 = data[:, 7]
y_ = data[:, -1]

X_ = np.column_stack((x_6, x_8))

# Calculate Mutual Information
mi_value_ = mutual_information_probability_space(X_, y_, bins=10)

print("\nMutual Information with own code for I(X; Y) for x=[x6, x8] and label y:", mi_value_)

#c) Calculate the MI with some clustering-based method (with some existing library, e.g. [ β ] )

# Calculate Mutual Information using clustering-based method (k-NN)
mi_value1 = mutual_info_classif(X, y, discrete_features=False, n_neighbors=5)

# Summing MI for each feature
mi_total = np.sum(mi_value1)

print("\nMutual Information I(X; Y) for x=[x6, x8] and label y (k-NN based):", mi_total)
```

```python
# 4) a)
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_selection import mutual_info_classif
import numpy as np
from itertools import combinations

data = np.loadtxt('/content/XY.txt', skiprows=1)

# Separate features and label
X_all = data[:, :-1]
y = data[:, -1]

# Number of features
n_features = X_all.shape[1]

# Initialize variables to store the best feature pair and max MI
best_pair = None
max_mi = -np.inf

# Iterate over all pairs of features
for i, j in combinations(range(n_features), 2):
    # Select the current pair of features
    X_pair = X_all[:, [i, j]]

    # Calculate MI between the pair of features and the label
    mi_value = np.sum(mutual_info_classif(X_pair, y, discrete_features=False, n_neighbors=5))

    # Update the best pair if this MI is the highest so far
    if mi_value > max_mi:
        max_mi = mi_value
        best_pair = (i, j)

print(f"The best feature pair for predicting y is: x[{best_pair[0]}] and x[{best_pair[1]}]")
print(f"Maximum Mutual Information I(y; [x[{best_pair[0]}], x[{best_pair[1]}]]) =", max_mi)

# c)
import numpy as np
from itertools import combinations
from sklearn.cluster import KMeans
from sklearn.metrics import mutual_info_score
from sklearn.preprocessing import KBinsDiscretizer

# Function to compute MI based on clustering
def compute_mi_clustering(X, y, n_clusters=10):
    # Apply KMeans clustering to the features X and the label y
    kmeans_X = KMeans(n_clusters=n_clusters, random_state=0).fit(X)
    kmeans_y = KMeans(n_clusters=n_clusters, random_state=0).fit(y.reshape(-1, 1))  # Reshape y to fit

    # Get the cluster assignments for the features and label
    X_labels = kmeans_X.labels_
    y_labels = kmeans_y.labels_

    # Compute the MI between the feature clusters and the label clusters
    mi_value = mutual_info_score(X_labels, y_labels)

    return mi_value

# Initialize variables to store the best feature pair and max MI
best_pair = None
max_mi = -np.inf

# Iterate over all pairs of features
for i, j in combinations(range(n_features), 2):
    # Select the current pair of features
    X_pair = X_all[:, [i, j]]

    # Calculate MI using clustering between the pair of features and the label
    mi_value = compute_mi_clustering(X_pair, y, n_clusters=10)

    # Update the best pair if this MI is the highest so far
    if mi_value > max_mi:
        max_mi = mi_value
        best_pair = (i, j)

# Display the best feature pair and the corresponding MI value
print(f"\nThe best feature pair for predicting y using clustering is: x[{best_pair[0]}] and x[{best_pair[1]}]")
print(f"Maximum Mutual Information I(y; [x[{best_pair[0]}], x[{best_pair[1]}]]) (Clustering-based) =", max_mi)
```
